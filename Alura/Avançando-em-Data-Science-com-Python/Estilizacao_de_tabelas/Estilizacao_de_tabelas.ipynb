# Missão 1: Time de vendas

Precisamos construir uma visualização para indicar os **10 maiores clientes por total de venda**.

Com isso, a equipe de vendas pode pensar em estratégias para manter esses clientes comprando em nossa loja.

```python
df_clientes_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
df_clientes_vendas= df_clientes_vendas.reset_index()
df_clientes_vendas
```

```python
df_clientes_vendas.columns = ['Cliente', 'Vendas']
df_clientes_vendas['Ranque']= df_clientes_vendas.index + 1
df_clientes_vendas.set_index('Ranque', inplace=True)
df_clientes_vendas
```

```python
s = df_clientes_vendas.style
s
```

```python
s.format({'Vendas': 'R$ {:,.2f}'})
s
```

# Missão 2: Time comercial

Precisamos construir uma visualização que permita associar o **total ganho em vendas e o total lucrado por tipo de produto**.

Isso vai permitir que o time comercial possa focar em estratégias para alavancar as rendas que ganham com cada tipo de produto vendido na loja.

```python
df_venda_lucro = df.groupby(['tipo_produto'])[['vendas', 'lucro']].sum()
df_venda_lucro.index.name = 'Tipo de produto'
df_venda_lucro
```

```python
estilo_produto = df_venda_lucro.style
estilo_produto
```

```python
estilo_produto.format('R$ {:,.2f}')\
.highlight_max(color='lightgreen')\
.highlight_min(color='#cd4f39')
estilo_produto
```

```python
estilo_produto.format('R$ {:,.2f}')\
.background_gradient(cmap='Blues')
estilo_produto
```

```python
cabecalho={
    'selector': 'th',
    'props': 'font-weight:bold ; font-family:Arial ; text-align: capitalize'
}
estilo_produto.set_table_styles([cabecalho],overwrite=False)
estilo_produto
```

# Missão 3: Time de logística

Precisamos construir uma visualização que permita entender a **distribuição de pedidos por região do Brasil**.

Isso vai permitir que o time logística entenda onde é mais interessante focar material e pessoas.

```python
df_regiao = pd.DataFrame(df['regiao'].value_counts())
df_regiao.columns = ['N° de pedidos']
df_regiao.index.name = 'Região'
df_regiao
```

```python
import numpy as np
```

```python
porcentagem = df_regiao['N° de pedidos'].to_numpy()
porcentagem = 100* porcentagem/porcentagem.sum()
porcentagem
```

```python
df_regiao['Porcentagem regiao'] = porcentagem
df_regiao
```

```python
estilo_regiao = df_regiao.style
estilo_regiao
```

```python
cabecalho={
    'selector': 'th',
    'props': 'font-weight:bold ; font-family:Arial ; text-align: rigth ; '#background-color:#fff'
}
celulas={
    'selector': 'tp',
    'props': 'background-color:#fff'
}
estilo_regiao.set_table_styles([cabecalho,celulas])
estilo_regiao
```

```python
estilo_regiao.format({'Porcentagem regiao': '{:.2f}%'})\
.bar(subset=['Porcentagem regiao'],vmin=0,vmax=100.0,color='#9CD33B')
estilo_regiao
```

# Missão 4: Time de logística

Novamente o time de logística deseja repensar sua estrutura e organização de fornecimento de produtos e para isso, foi pensada em uma visualização que permita identificar padrão de **quantidade de produtos solicitados por mês por departamento**.

```python
df_meses = df.copy()
df_meses = df_meses.sort_values('data_pedido')
df_meses
```

```python
df_meses['meses'] = df_meses['data_pedido'].dt.strftime('%Y - %b')
df_meses = df_meses.reset_index(drop=True)
df_meses
```

```python
vendas_mensal = df_meses.pivot_table(index='departamento',columns='meses',values='quantidade',aggfunc='sum',sort = False)
vendas_mensal
```

```python
estilo_mensal = vendas_mensal.style
estilo_mensal
```

```python
estilo_mensal.set_sticky(axis='index')
estilo_mensal
```

```python
clunas= {
    'selector': '.col_heading',
    'props': 'font-weight:normal ; font-family:Arial ; text-align: rigth ; '#background-color:#fff'
}
tabelas={
    'selector': 'td,th',
    'props':  'text-align: left ; '#background-color:#fff'
}
index={
    'selector': '.index_name',
    'props': 'font-weight:bold ; font-family:Arial ; text-align: rigth ; '
}
estilo_mensal.set_table_styles([clunas,tabelas,index],overwrite=False)
estilo_mensal
```

# Missão 5: Relatório de performance

O relatório de performance é um documento muito importante para a empresa. Nele, todos que trabalham na loja podem entender como anda a empresa. Por conta disso, tabelas bem construídas são essenciais para mostrar os dados da melhor forma.

Portanto, precisamos construir uma visualização que permita entender a **relação entre tipos de clientes e modo de envio de produtos de acordo com as vendas**. Com isso, será possível destacar no relatório qual o modo de envio e quais tipos de modo de envio mais geraram ganho em vendas.

```python
df_client =df.pivot_table(index='segmento_cliente',columns='modo_envio',values='vendas',aggfunc='sum')
df_client
```

```python
df_client['Total']=df_client.sum(axis=1)
df_client.loc['Total']=df_client.sum()
df_client
```

```python
compra_cliente = df_client.style.format('{:,.2f}')\

compra_cliente
```

```python
tabela = {
    'selector': 'td,th:not(.index_name)',
    'props': 'font-weight: normal; font-family: Arial; text-align: center; background-color: white'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: normal; text-align: rigth; font-style: italic; color: #696969'
}
compra_cliente.set_table_styles([tabela,index])
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    {
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }],
    'B2B': [{
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    {
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }]

}, overwrite = False, axis=1)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #D8D8D8'
    }]
}, overwrite=False, axis=0)

cores_coluna = pd.DataFrame(['false', 'true', 'false'], index= df_client['Total'].index, columns=['Total'])
compra_cliente.set_td_classes(cores_coluna)
```

```python
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mês)']= dados['Valor']+dados['Condominio']
```

```python
dados
```

```python
dados['Valor_por_ano)']= dados['Valor_por_mês)']*12 + dados['IPTU']
```

```python
dados.head()
```

```python
dados['Descrição']= dados['Tipo'] + ' em ' + dados['Bairro']
```

```python
dados
```

```python
dados['Descrição']= dados['Tipo'] + ' em ' + dados['Bairro'] + 'com' + dados['Quartos'] + 'quarto(s)'\
+ ' e ' + dados['Vagas'] + 'Vaga(s) de garagem'
```

```python
dados['Descrição']= dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) '\
+ ' e ' + dados['Vagas'].astype(str) + ' Vaga(s) de garagem'
```

```python
dados.head()
```

```python
dados['Possui_suite']= dados['Suites'].apply(lambda x: 'Sim' if x>0 else 'Não')
```
